```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy="most_frequent")
empl["City"]=imp.fit_transform(empl["City"].values.reshape(-1,1))
empl
```

```python
le=LabelEncoder()
list=["City"] 
for val in list:
    empl[val]=le.fit_transform(empl[val].astype(str))
```

```python
empl
```

# SPLIT DATA INTO X AND Y 

```python
x=empl.drop("Salary",axis=1)
```

```python
x
```

```python
y=empl["Salary"]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.21,random_state=42)
```

# APPLYING MODEL

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.score(x_train,y_train)
```

```python
pred=lm.predict(x_test)
pred
```

```python
empl1=pd.DataFrame({"Actual":y_test,"Predicted":pred})
```

```python
empl1
```

```python
print("Mean Absolute Error:", metrics.mean_absolute_error(y_test,pred))
print("Mean Squared Error:", metrics.mean_squared_error(y_test,pred))
print("Root Mean Squared Error:", np.sqrt(metrics.mean_squared_error(y_test,pred)))
```
