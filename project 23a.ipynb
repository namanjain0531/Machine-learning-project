# SPLIT THE DATA INTO TRAINING AND TESTING PHASE AND APPLY MODEL

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.21,random_state=10)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

# MAKE PIPELINE AND APPLY MODEL

```python
from sklearn.pipeline import make_pipeline
```

```python

from sklearn.ensemble import RandomForestRegressor
```

```python
pipeline=make_pipeline(RandomForestRegressor(n_estimators=100))
```

```python
pipeline.get_params()
```

```python
hyperparameters = { 'randomforestregressor__max_features' : ['auto','sqrt'],
                  'randomforestregressor__max_depth': [None,1,2,3]}
```

# USING GRIDSEARCHCV MODEL

```python
from sklearn.model_selection import GridSearchCV
```

```python
gsv=GridSearchCV(pipeline,hyperparameters,cv=10)
gsv.fit(x_train,y_train)
ypred=gsv.predict(x_test)
print(r2_score(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
print(lr.score(x_train,y_train))
ypred=lr.predict(x_test)
print(r2_score(y_test,ypred))
print(mean_squared_error(y_test,ypred))
```

```python
#save the model
import warnings
warnings.filterwarnings('ignore')
from sklearn.externals import joblib
joblib.dump(gsv,"wine.pkl")
```
